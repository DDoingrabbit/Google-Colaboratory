## **iloc[ ] vs loc[ ]**
**위치 기반 인덱싱 vs 명칭 기반 인덱싱**


*   위치 기반 인덱싱 : 행, 열의 위치 좌표에만 의존
*   명칭 기반 인덱싱 : 엔덱스, 칼럼명으로 데이터 접근



```python
data = {'Name' : ['Kimin','Cheolsoo','Younghee','Minsoo'],
        'Year' : [1988, 1992, 1987, 1994],
        'Gender' : ['Male','Male','Female','Male']}
data_df = pd.DataFrame(data, index = ['one','two','three','four'])
```

```python
data_df
```

## **groupby()**


*   파라미터 by에 입력된 칼럼으로 groupby 된다
*   DataFrameGroupBy라는 DataFrame 객체 반환
*   특정 칼럼들을 추출가능
*   Aggregation 함수를 적용하기 전에 해당 칼럼들을 필터링 해주어야 한다





```python
titanic_groupby = titanic_df.groupby(by = 'Pclass')[['PassengerId', 'Survived']].count()
print(type(titanic_df.groupby(by = 'Pclass')[['PassengerId', 'Survived']]))
titanic_groupby
```

DataFrame에서 여러개의 aggregation함수를 사용해야 할 경우, DataFrameGroupBy 객체의 agg()내에 인자로 넣어서 사용하면 된다

```python
titanic_df.groupby('Pclass')['Age'].agg([max, min])
```

SQL에서 max(Age), sum(Sibsp), avg(Fare) frokm titanic_table group by Pclass 처럼 사용하려면 다음과 같이 agg() 인자를 딕셔너리 형태로 줄 수 있음

```python
agg_format = {'Age' : 'max' ,'SibSp' : 'sum', 'Fare' : 'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

# **결손 데이터 처리하기**
## **결손 데이터 여부 확인 : isna()**
칼럼 값이 NaN이면 True를 반환한다. sum()을 이용하여 결손 데이터 갯수도 구할 수 있다

```python
titanic_df.isna().sum()
```

## **결손 데이터 처리 : fillna()**
fillna()를 사용하여 NaN를 다른 값으로 대체할 수 있다.

이 예제에서는 결손 데이터를 다음과 같이 처리

*   'Cabin'의 NaN : 'COOO'
*   'Age'의 NaN : 'Age'의 평균
*   'Embarked'의 NaN : 'S'





```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('s')
```

```python
titanic_df.isna().sum()
```

# **데이터 가공 : apply lambda**
Pandas는 apply 함수에 lambda식을 결합하여 DataFrame이나 Series의 레코드별로 데이터를 가공할 수 있다. 

다음과 같이 Child와 Adult를 나눌 경우 사용하면 좋다

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult')
titanic_df['Child_Adult'].value_counts()
```
